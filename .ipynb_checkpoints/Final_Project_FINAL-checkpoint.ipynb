### <ins>Introduction</ins>
Macronutrients, such as phosphorus, nitrogen, and silicon, are critical components of seawater, playing essential roles in marine ecosystems. Phosphorus and nitrogen support the growth and function of phytoplankton, which form the base of the marine food web and contribute nearly half of global primary production (Moore et al., 2013; Field et al., 1998). Through photosynthesis, phytoplankton influence the global carbon cycle by sequestering atmospheric carbon dioxide. Dissolved phosphorus, primarily in the form of phosphate (PO₄), is utilized by phytoplankton for DNA and RNA synthesis and energy transfer (Paytan and McLaughlin, 2007). Phosphate availability directly regulates phytoplankton productivity. Diatoms, a dominant group of phytoplankton, account for up to 40% of oceanic primary production (Mann, 1999) and require dissolved silicon, in the form of silicic acid, to build their cell walls (Yool et al., 2003). Silicate availability influences diatom growth rates (Conley et al., 1997). Thus, nutrient cycling in the ocean is tightly coupled to the global carbon cycle and climate system (Paytan and McLaughlin, 2007).

Expanding the temporal and spatial resolution of nutrient measurements is critical for refining nutrient cycling models, reducing uncertainties, and improving predictions of global biogeochemical processes (Paytan and McLaughlin, 2007; Tréguer et al., 2013). However, the significant spatial and temporal variability of macronutrients, coupled with logistical constraints of ship-based sampling, presents challenges for continuous monitoring. Automated, in-situ sensors offer a promising solution, enabling broader and more consistent nutrient observations.

Current sensors for phosphate and silicate measurement, although improved, still face limitations. Phosphate sensors typically rely on actuator-sensor systems (Chen et al., 2023), while recent in-situ electrochemical sensors for silicate require approximately 45 minutes per sample, limiting their resolution (Barus et al., 2018; Legrand et al., 2021). In contrast, nitrate sensors deployed on profiling floats can operate autonomously for years, detecting daily to seasonal fluctuations using reagent-free ultraviolet spectrophotometers (Johnson et al., 2013; Sarmiento et al., 2023). Expanding similar autonomous technologies to phosphate and silicate monitoring is vital for advancing understanding of marine biogeochemical cycles.

Programmable flow injection (pFI) represents a major step toward achieving efficient, autonomous macronutrient analysis. Originally developed for trace metal analysis (Grand et al., 2019), pFI evolved from earlier techniques such as segmented flow analysis and flow injection analysis (Worsfold et al., 2013; Ruzicka and Chocholous, 2024). The pFI system is a compact, microfluidic benchtop instrument that uses two milliGAT bidirectional pumps and a computer-controlled lab-on-valve manifold to move reagents and samples through holding coils and a flow cell (Figure 1; Appendix 1). Solutions are aspirated into temperature-controlled holding coils to allow chemical equilibrium before detection. Initial optimization of pFI for nutrient analysis focused on refining chemical reactions for phosphate and silicate detection (Hatta et al., 2021, 2023).

Figure 1: Schematic of the pFI instrument (Hatta et al., 2023).



For long-term autonomous deployment, maintaining reagent and standard stability is crucial to ensure measurement accuracy. Optimization efforts for pFI deployments focused on reagent concentrations, reagent-to-sample ratios, and mitigating factors that cause standard degradation (Lebrec and Grand, 2024). Silicate standards are particularly sensitive to pH, with significant concentration losses occurring at acidities above 11.2 mM H⁺. Phosphate standards degrade over time when unacidified, primarily due to the adsorption of phosphorus anions onto container walls. Acidifying phosphate standards to 13.3 mM H⁺ mitigates this issue without affecting analytical accuracy. Biological uptake may also contribute to phosphate loss. Although biocides can inhibit microbial activity, common options like mercuric chloride present environmental risks and potential measurement interference (Ge et al., 2022; Kattner, 1999).

Building on the work of Lebrec and Grand (2024), this study develops preservation strategies for phosphate and silicate working standards over a 35-day period. Four treatments were applied: a control (Milli-Q water), acid addition, biocide addition, and a combination of acid and biocide. Treatments were designed to distinguish between chemical and biological drivers of standard instability. For phosphate standards, the study evaluates whether concentration declines stem from adsorption or microbial consumption. For silicate standards, the impact of acidification on stability is assessed. Additionally, the study tests the combined behavior of phosphate and silicate standards, and the effect of adding ProClin-300, a low-toxicity antimicrobial biocide. By analyzing concentration changes over 35 days, this work aims to optimize long-term, autonomous deployments of pFI analyzers for accurate macronutrient measurements in marine environments.

### <ins>Materials and Methods</ins>

<ins>Treatment Development</ins>

Silicate is highly sensitive to pH changes, with solubility increasing under acidified conditions. To determine the optimal acid concentration for preserving phosphate while maintaining silicate stability, a preliminary experiment was conducted. Four sample types were prepared by spiking 20 µL of phosphate (PO₄) standard and 200 µL of silicate (Si) standard into Milli-Q water, 1.5 mM HCl, 3.0 mM HCl, or 6.0 mM HCl solutions (Table 1). Acid solutions were prepared in 250 mL glass flasks by diluting concentrated HCl with Milli-Q water.

Table 1. Volumes of HCl added to each solution.



Concentrations were determined from absorbance measurements and analyzed using a one-tailed two-sample t-test. Results indicated that 6.0 mM HCl significantly reduced silicate concentration (Table 2), while 1.5 mM and 3.0 mM HCl had no significant effect. Therefore, 3.0 mM HCl was selected as the optimal acid concentration. 

Table 2. Mean Si concentrations (µM ± combined error) and t-test results.




For microbial control, ProClin-300 (3% solution) was used at a final concentration of 12 ppm, within the manufacturer’s recommended range of 6–20 ppm.

<ins>Sample Preparation</ins>

All solutions were prepared using high-purity Milli-Q water (18.2 MΩ·cm). Stock standards were made from salts dried at 100°C for 24 hours and stored in a desiccator.

Phosphate stock (4998 µM): 0.678 g KH₂PO₄ (Baker, Lot 39548) dissolved in 1 L Milli-Q water.
Silicate stock (10,255 µM): Na₂SiF₆ dissolved in 500 mL Milli-Q water (plastic container).

Samples were prepared in 250 mL HDPE bottles, acid-cleaned in 10% HCl for 24 hours. Four treatments were applied (Table 3), with each sample adjusted to a final volume of 250 mL with Milli-Q water. To reach final concentrations of 2 µM PO₄ and 41 µM Si, 0.1 mL of phosphate stock and 2.0 mL of silicate stock were added per bottle.

Table 3. Sample compositions by treatment group.




<ins>Reagent Preparation</ins>

Reagents were prepared fresh in 50 mL Falcon tubes:

Ascorbic acid reagent: 1.1 g ascorbic acid and 0.125 g SDS in Milli-Q water, adjusted to 50 mL.
Oxalic acid reagent: 1.0 g oxalic acid in 50 mL Milli-Q water.
Acidified molybdate for silicate: 0.25 g ammonium molybdate tetrahydrate dissolved in 20 mL Milli-Q water, with 750 µL 6.02 M sulfuric acid and 0.125 g SDS, diluted to 50 mL.
Acidified molybdate for phosphate: 0.20 g ammonium molybdate tetrahydrate in 20 mL Milli-Q water, with 1.5 mL potassium antimony tartrate stock, 11 mL 5.0 M HCl, and 0.125 g SDS, diluted to 50 mL.
SDS carrier: 0.25% w/v SDS (2.5 g SDS in 1 L Milli-Q water).

<ins>Instrumentation</ins>

Samples and reagents were loaded into Sarstedt tubes and connected to the miniSIA-2 analyzer (GlobalFIA), equipped with two milliGAT pumps, a thermostated 10 cm flow cell, a tungsten-halogen light source, and an Ocean Optics spectrophotometer. Sample analysis was conducted using FLO ZF 5.3.0451 software. To minimize instrumental drift, samples were analyzed in randomized order and each measurement was performed in triplicate.

### Results (including code, figures and descriptions)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
from scipy import stats
from statsmodels.stats.power import TTestPower
import sys
sys.path.append("py")
import chem_calcs as cc
```

```python
#fixing pFI .txt files
files = ['DAY0.txt', 'DAY7.txt', 'DAY14.txt', 'DAY21.txt', 'DAY35.txt', 'DAY96.txt']

#fix sample names since i manually named them all differently
Si_fixes = {"A2": "A2_Si", "A3": "A3_Si", "A4": "A4_Si", "B2": "B2_Si", "B3": "B3_Si", "B4": "B4_Si", "C2": "C2_Si", "C3": "C3_Si", "C4": "C4_Si", "D1": "D1_Si", "D2": "D2_Si"}
P_fixes = {"A1_P": "A1_PO4", "A3_P": "A3_PO4", "A4_P": "A4_PO4", "B1_P": "B1_PO4", "B3_P": "B3_PO4", "B4_P": "B4_PO4", "C1_P": "C1_PO4", "C3_P": "C3_PO4", "C4-P": "C4_PO4", "D1_P": "D1_PO4", "D2_P": "D2_PO4"}

Si_samples = list(Si_fixes.values())
P_samples = list(P_fixes.values())

#calcultaing Si standard concentration input (concentration, spike, new concentation, new bottle volume)
Si_conc = cc.conc_to_dil(10255, 220, None, 50 * 1000)
P_conc = cc.conc_to_dil(4997.89, 40, None, 50*1000)

#calculating pFI calibration dilutions that will be used for calibration curve
dilutions = np.array([0, (600-450)/600, (600-300)/600, (600-150)/600, (600-50)/600])
dil_series = np.repeat(dilutions, 3)
Si_cali_concs = dil_series * Si_conc
P_cali_concs = dil_series * P_conc
```

```python
#Si calculations
#creating data frame
Si_df = []
#for each file take the number from the .txt file name for the day
for file in files:
    day = int(file.replace('DAY', '').replace('.txt', ''))
    #filter txt file for needed information only
    df = pd.read_csv("Data/" + file, sep="\t")
    df = df[['Name', 'AVE810_1050', 'Sequence Name']].copy()
    df['Name'] = df['Name'].replace(Si_fixes)
    
    #get the specific Si absorbance of each sample
    Si_cali_abs = df.loc[df['Sequence Name'] == "AutoCalibration_Silicate", 'AVE810_1050'].to_numpy()

    #create calibration curve for the day
    fit = stats.linregress(Si_cali_concs, Si_cali_abs)
    slope = fit.slope
    yint = fit.intercept
    slope_err = fit.stderr
    yint_err = fit.intercept_stderr
    #calculating uncertainity variables to calculate error
    predicted = slope * Si_cali_concs + yint
    residuals = Si_cali_abs - predicted
    Sxo = np.sqrt(np.sum(residuals**2) / (len(Si_cali_concs) - 2))

    #calculate Si concentrations from the mean absorbances (3 replicates)
    for sample in Si_samples:
        avg_abs = df.loc[df['Name'] == sample, 'AVE810_1050'].mean()
        conc, u_combined = cc.calc_Si_conc_and_uncert(avg_abs, yint, slope, Sxo, slope_err, yint_err)
        Si_df.append({'Sample': sample, 'Day': day, 'Concentration (uM)': conc, 'Uncertainty': u_combined})

#fill a final data frame
Si_results = pd.DataFrame(Si_df)
Si_results = Si_results.sort_values(by=['Sample', 'Day']).reset_index(drop=True)

print(Si_results)
```

```python
#P calculations
#creating data frame
P_df = []

#for each file take the number from the .txt file name for the day
for file in files:
    day = int(file.replace('DAY', '').replace('.txt', ''))
    #filter txt file for needed information
    df = pd.read_csv("Data/" + file, sep="\t")
    df = df[['Name', 'AVE880_1050', 'Sequence Name']].copy()
    df['Name'] = df['Name'].replace(P_fixes)

    #get absorbance of each sample at specific wavelength
    P_cali_abs = df.loc[df['Sequence Name'] == "Autocalibration_Phosphate", 'AVE880_1050'].to_numpy()
    #create calibration curve for the day
    fit = np.polyfit(P_cali_concs, P_cali_abs, 2)
    a, b, c = fit
    
    #get uncertainity variables to calculate error
    P_predicted_abs = a * P_cali_concs**2 + b * P_cali_concs + c

    #Residuals: observed - predicted
    P_residuals = P_cali_abs - P_predicted_abs

    #Sxo for PO4
    n_P = len(P_cali_concs)
    Sxo_P = np.sqrt(np.sum(P_residuals**2) / (n_P - 3))

    #calculate Si concentrations from the mean absorbances
    for sample in P_samples:
        avg_abs = df.loc[df['Name'] == sample, 'AVE880_1050'].mean()

        if avg_abs == avg_abs:  #skips NaN values in .txt file
            conc, u_combined = cc.calc_PO4_conc_and_uncert(avg_abs, a, b, c, Sxo_P)
            P_df.append({'Sample': sample, 'Day': day, 'Concentration (uM)': conc, 'Uncertainty': u_combined})

#concentration data frame
P_results = pd.DataFrame(P_df)
P_results = P_results.sort_values(by=['Sample', 'Day']).reset_index(drop=True)

print(P_results)
```

```python
#graphs of PO4 concentration over time
P_results['Row'] = P_results['Sample'].str[0]

#label and assign colors to treatment groups
labels = {'A': 'Acid Treatment', 'B': 'Biocide Treatment', 'C': 'Control', 'D': 'Acid and Biocide Treatment'}
colormaps = {'A': plt.cm.Reds, 'B': plt.cm.Greens, 'C': plt.cm.Blues, 'D': plt.cm.copper}

#graph
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

#each treatment
for i, row in enumerate(['A', 'B', 'C', 'D']):
    ax = axes[i]
    data = P_results[P_results['Row'] == row]
    samples = data['Sample'].unique()
    cmap = colormaps[row]
    sample_colors = dict(zip(samples, cmap(np.linspace(0.3, 0.8, len(samples)))))

    #each sample with errors
    for sample in samples:
        s = data[data['Sample'] == sample]
        color = sample_colors[sample]
        ax.errorbar(s['Day'], s['Concentration (uM)'], yerr=s['Uncertainty'], fmt='.', color=color, alpha=0.6, capsize=2)

        #linear regressions
        x = s['Day']
        y = s['Concentration (uM)']
        slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
        if p_value < 0.05:
            print(f"{sample} is significant: p = {p_value:.4f}")
        x_fit = np.linspace(x.min(), x.max(), 100)
        y_fit = slope * x_fit + intercept
        ax.plot(x_fit, y_fit, '-', color=color, label=f'{sample} (p={p_value:.3f})')

    ax.set_title(labels[row])
    ax.set_xlabel('Time (Days since treatments)')
    ax.set_ylabel('PO4 (µM)')
    ax.legend(fontsize=8, loc='upper right')

plt.tight_layout()
plt.show()
```

```python
#Si graphs of concentration over time
Si_results['Row'] = Si_results['Sample'].str[0]

#label and assign colors to treatment groups
labels = {'A': 'Acid Treatment', 'B': 'Biocide Treatment', 'C': 'Control', 'D': 'Acid and Biocide Treatment'}
colormaps = {'A': plt.cm.Reds, 'B': plt.cm.Greens, 'C': plt.cm.Blues, 'D': plt.cm.copper}

#graph
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

#each treatment
for i, row in enumerate(['A', 'B', 'C', 'D']):
    ax = axes[i]
    data = Si_results[Si_results['Row'] == row]
    samples = data['Sample'].unique()
    cmap = colormaps[row]
    sample_colors = dict(zip(samples, cmap(np.linspace(0.3, 0.8, len(samples)))))

    #each sample with errors
    for sample in samples:
        s = data[data['Sample'] == sample]
        color = sample_colors[sample]
        ax.errorbar(s['Day'], s['Concentration (uM)'], yerr=s['Uncertainty'], fmt='.', color=color, alpha=0.6, capsize=2)

        #linear regressions
        x = s['Day']
        y = s['Concentration (uM)']
        slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
        if p_value < 0.05:
            print(f"{sample} is significant: p = {p_value:.4f}")
        x_fit = np.linspace(x.min(), x.max(), 100)
        y_fit = slope * x_fit + intercept
        ax.plot(x_fit, y_fit, '-', color=color, label=f'{sample} (p={p_value:.3f})')

    ax.set_title(labels[row])
    ax.set_xlabel('Time (Days since treatment)')
    ax.set_ylabel('Si (µM)')
    ax.legend(fontsize=8, loc='lower right')

plt.tight_layout()
plt.show()
```

```python
#paired t-test comparing day 0 and day 35 for each sample
start_day = 0
end_day = 35

print('Silicate Paired T-test Results Day 0 to Day 35:')
Si_results['Row'] = Si_results['Sample'].str[0]

for treatment in ['A', 'B', 'C', 'D']:
    df_start = Si_results[(Si_results['Day'] == start_day) & (Si_results['Row'] == treatment)]
    df_end = Si_results[(Si_results['Day'] == end_day) & (Si_results['Row'] == treatment)]
    
    paired_df = pd.merge(df_start, df_end, on='Sample', suffixes=('_start', '_end'))
    t_stat, p_val = stats.ttest_rel(paired_df['Concentration (uM)_start'], paired_df['Concentration (uM)_end'])
    #effect size and power analysis
    diff = paired_df['Concentration (uM)_end'] - paired_df['Concentration (uM)_start']
    cohen_d = diff.mean() / diff.std(ddof=1)
    power = TTestPower().power(effect_size=cohen_d, nobs=len(paired_df), alpha=0.05)

    print(f"Treatment {treatment}: t = {t_stat:.4f}, p = {p_val:.4f}, d = {cohen_d:.3f}, power = {power:.3f}")
print('Phosphate Paired T-test Results Day 0 to Day 35:')
P_results['Row'] = P_results['Sample'].str[0]

for treatment in ['A', 'B', 'C', 'D']:
    df_start = P_results[(P_results['Day'] == start_day) & (P_results['Row'] == treatment)]
    df_end = P_results[(P_results['Day'] == end_day) & (P_results['Row'] == treatment)]
    
    paired_df = pd.merge(df_start, df_end, on='Sample', suffixes=('_start', '_end'))
    t_stat, p_val = stats.ttest_rel(paired_df['Concentration (uM)_start'], paired_df['Concentration (uM)_end'])
    #effect size and power analysis
    diff = paired_df['Concentration (uM)_end'] - paired_df['Concentration (uM)_start']
    cohen_d = diff.mean() / diff.std(ddof=1)
    power = TTestPower().power(effect_size=cohen_d, nobs=len(paired_df), alpha=0.05)

    print(f"Treatment {treatment}: t = {t_stat:.4f}, p = {p_val:.4f}, d = {cohen_d:.3f}, power = {power:.3f}")
```

## Discussion
The phosphate and silicate calculations utilize calibration curves. The calibration curves were compared to previously created calibration curves and matched. The concentrations calculated match the concentrations previously calculated. The uncertainty is higher and most likely more accurate. These are great outcomes for this project, showing that a .txt file straight from the software program can accurately calculate the concentration of a sample and the uncertainty associated with the measurement.

The linear regressions performed on the treatments showed no overall treatment concentration change over time for phosphate or silicate. The graphs include the p-values from the linear regressions. The graphs visualize each treatment's samples' calculated concentrations over time with the calculated error.

The paired t-test performed compared the concentrations of each treatment from day 0 to day 35, so one sample over time. The only sample p-value that was significant (< 0.05) is the Si treatment control. This is an interesting result because graphically it doesn’t seem like an outlier, nor does it show a decrease, as the t-value shows -17. However, the effect size is large (too large for the linear regression), so the power analysis shows NaN. This sample could have some issues or outliers that need to be dealt with.

The power test (testing for the chance that you could detect an effect that isn’t there) shows very low power, so the sample size is too small to reliably detect a change. For the power to be 0.8 (an 80% chance of detecting a significant effect when there is one), the sample size would need to be around 15, so more replicates are needed.

The paired t-test and linear regression are good for showing concentration changes over time visually and overall; however, more statistical analysis could explain each sample's change over time and help determine the best treatment to utilize. The use of a repeated measures ANOVA would be useful too, since I am measuring the same samples' concentrations over time. This could be implemented using the three replicate measured absorbances per sample per measurement day and calculating concentrations for all of them, and then running the repeated measures ANOVA on those concentrations. However, the sphericity assumption might not be met, and the test is sensitive to outliers when sample sizes are small, so I need to get the outliers taken care of. Then I could do this, but I would sometimes have only two samples, which would not be enough. As an alternative, non-parametric tests such as the Kruskal-Wallis test (for comparing more than two groups) or pairwise Mann-Whitney U tests (for comparing between two timepoints or treatments) could be applied. These tests do not assume normality or sphericity, making them more robust for small or uneven sample sizes, though they may be less powerful than parametric tests when assumptions are met.

## Conclusion and Ideas for future work

This project streamlined the concentration calculation process, and now .txt files can be taken from the software and the concentrations and errors can be calculated using this code. There are still improvements to be made that will benefit future use. Sometimes with pFI, there are bubbles in the tubing that cause the absorbances to decrease enormously. Creating a code to use when you upload the file before processing that finds the outliers and removes them would be highly advantageous. However, this may lead to even fewer data samples (only two replicates per sample), which is not great for statistical analysis. I will attempt to utilize Dixon’s Q-test (this is what is utilized in chemical oceanography to my knowledge): (abs(suspected value – nearest value)) / (largest value – smallest value).

Another improvement that I intend to make is in the naming of samples within the software. The computer program records each sample, and you manually enter a name for each sample. This created a lot of issues for my data processing, since each sample had a letter from its treatment group, a number, and Si or P. Going into an experiment with a set name for each sample and entering it the same each day will make the .txt files more uniform and easier to manipulate.
