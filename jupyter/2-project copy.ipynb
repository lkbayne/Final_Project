### <ins>Introduction</ins>
Macronutrients, such as phosphorus, nitrogen, and silicon, are critical components of seawater, playing essential roles in marine ecosystems. Phosphorus and nitrogen support the growth and function of phytoplankton, which form the base of the marine food web and contribute nearly half of global primary production (Moore et al., 2013; Field et al., 1998). Through photosynthesis, phytoplankton influence the global carbon cycle by sequestering atmospheric carbon dioxide. Dissolved phosphorus, primarily in the form of phosphate (PO₄), is utilized by phytoplankton for DNA and RNA synthesis and energy transfer (Paytan and McLaughlin, 2007). Phosphate availability directly regulates phytoplankton productivity. Diatoms, a dominant group of phytoplankton, account for up to 40% of oceanic primary production (Mann, 1999) and require dissolved silicon, in the form of silicic acid, to build their cell walls (Yool et al., 2003). Silicate availability influences diatom growth rates (Conley et al., 1997). Thus, nutrient cycling in the ocean is tightly coupled to the global carbon cycle and climate system (Paytan and McLaughlin, 2007).

Expanding the temporal and spatial resolution of nutrient measurements is critical for refining nutrient cycling models, reducing uncertainties, and improving predictions of global biogeochemical processes (Paytan and McLaughlin, 2007; Tréguer et al., 2013). However, the significant spatial and temporal variability of macronutrients, coupled with logistical constraints of ship-based sampling, presents challenges for continuous monitoring. Automated, in-situ sensors offer a promising solution, enabling broader and more consistent nutrient observations.

Current sensors for phosphate and silicate measurement, although improved, still face limitations. Phosphate sensors typically rely on actuator-sensor systems (Chen et al., 2023), while recent in-situ electrochemical sensors for silicate require approximately 45 minutes per sample, limiting their resolution (Barus et al., 2018; Legrand et al., 2021). In contrast, nitrate sensors deployed on profiling floats can operate autonomously for years, detecting daily to seasonal fluctuations using reagent-free ultraviolet spectrophotometers (Johnson et al., 2013; Sarmiento et al., 2023). Expanding similar autonomous technologies to phosphate and silicate monitoring is vital for advancing understanding of marine biogeochemical cycles.

Programmable flow injection (pFI) represents a major step toward achieving efficient, autonomous macronutrient analysis. Originally developed for trace metal analysis (Grand et al., 2019), pFI evolved from earlier techniques such as segmented flow analysis and flow injection analysis (Worsfold et al., 2013; Ruzicka and Chocholous, 2024). The pFI system is a compact, microfluidic benchtop instrument that uses two milliGAT bidirectional pumps and a computer-controlled lab-on-valve manifold to move reagents and samples through holding coils and a flow cell (Figure 1; Appendix 1). Solutions are aspirated into temperature-controlled holding coils to allow chemical equilibrium before detection. Initial optimization of pFI for nutrient analysis focused on refining chemical reactions for phosphate and silicate detection (Hatta et al., 2021, 2023).

Figure 1: Schematic of the pFI instrument (Hatta et al., 2023).



For long-term autonomous deployment, maintaining reagent and standard stability is crucial to ensure measurement accuracy. Optimization efforts for pFI deployments focused on reagent concentrations, reagent-to-sample ratios, and mitigating factors that cause standard degradation (Lebrec and Grand, 2024). Silicate standards are particularly sensitive to pH, with significant concentration losses occurring at acidities above 11.2 mM H⁺. Phosphate standards degrade over time when unacidified, primarily due to the adsorption of phosphorus anions onto container walls. Acidifying phosphate standards to 13.3 mM H⁺ mitigates this issue without affecting analytical accuracy. Biological uptake may also contribute to phosphate loss. Although biocides can inhibit microbial activity, common options like mercuric chloride present environmental risks and potential measurement interference (Ge et al., 2022; Kattner, 1999).

Building on the work of Lebrec and Grand (2024), this study develops preservation strategies for phosphate and silicate working standards over a 35-day period. Four treatments were applied: a control (Milli-Q water), acid addition, biocide addition, and a combination of acid and biocide. Treatments were designed to distinguish between chemical and biological drivers of standard instability. For phosphate standards, the study evaluates whether concentration declines stem from adsorption or microbial consumption. For silicate standards, the impact of acidification on stability is assessed. Additionally, the study tests the combined behavior of phosphate and silicate standards, and the effect of adding ProClin-300, a low-toxicity antimicrobial biocide. By analyzing concentration changes over 35 days, this work aims to optimize long-term, autonomous deployments of pFI analyzers for accurate macronutrient measurements in marine environments.

### <ins>Materials and Methods</ins>

<ins>Treatment Development</ins>

Silicate is highly sensitive to pH changes, with solubility increasing under acidified conditions. To determine the optimal acid concentration for preserving phosphate while maintaining silicate stability, a preliminary experiment was conducted. Four sample types were prepared by spiking 20 µL of phosphate (PO₄) standard and 200 µL of silicate (Si) standard into Milli-Q water, 1.5 mM HCl, 3.0 mM HCl, or 6.0 mM HCl solutions (Table 1). Acid solutions were prepared in 250 mL glass flasks by diluting concentrated HCl with Milli-Q water.

Table 1. Volumes of HCl added to each solution.



Concentrations were determined from absorbance measurements and analyzed using a one-tailed two-sample t-test. Results indicated that 6.0 mM HCl significantly reduced silicate concentration (Table 2), while 1.5 mM and 3.0 mM HCl had no significant effect. Therefore, 3.0 mM HCl was selected as the optimal acid concentration. 

Table 2. Mean Si concentrations (µM ± combined error) and t-test results.




For microbial control, ProClin-300 (3% solution) was used at a final concentration of 12 ppm, within the manufacturer’s recommended range of 6–20 ppm.

<ins>Sample Preparation</ins>

All solutions were prepared using high-purity Milli-Q water (18.2 MΩ·cm). Stock standards were made from salts dried at 100°C for 24 hours and stored in a desiccator.

Phosphate stock (4998 µM): 0.678 g KH₂PO₄ (Baker, Lot 39548) dissolved in 1 L Milli-Q water.
Silicate stock (10,255 µM): Na₂SiF₆ dissolved in 500 mL Milli-Q water (plastic container).

Samples were prepared in 250 mL HDPE bottles, acid-cleaned in 10% HCl for 24 hours. Four treatments were applied (Table 3), with each sample adjusted to a final volume of 250 mL with Milli-Q water. To reach final concentrations of 2 µM PO₄ and 41 µM Si, 0.1 mL of phosphate stock and 2.0 mL of silicate stock were added per bottle.

Table 3. Sample compositions by treatment group.




<ins>Reagent Preparation</ins>

Reagents were prepared fresh in 50 mL Falcon tubes:

Ascorbic acid reagent: 1.1 g ascorbic acid and 0.125 g SDS in Milli-Q water, adjusted to 50 mL.
Oxalic acid reagent: 1.0 g oxalic acid in 50 mL Milli-Q water.
Acidified molybdate for silicate: 0.25 g ammonium molybdate tetrahydrate dissolved in 20 mL Milli-Q water, with 750 µL 6.02 M sulfuric acid and 0.125 g SDS, diluted to 50 mL.
Acidified molybdate for phosphate: 0.20 g ammonium molybdate tetrahydrate in 20 mL Milli-Q water, with 1.5 mL potassium antimony tartrate stock, 11 mL 5.0 M HCl, and 0.125 g SDS, diluted to 50 mL.
SDS carrier: 0.25% w/v SDS (2.5 g SDS in 1 L Milli-Q water).

<ins>Instrumentation</ins>

Samples and reagents were loaded into Sarstedt tubes and connected to the miniSIA-2 analyzer (GlobalFIA), equipped with two milliGAT pumps, a thermostated 10 cm flow cell, a tungsten-halogen light source, and an Ocean Optics spectrophotometer. Sample analysis was conducted using FLO ZF 5.3.0451 software. To minimize instrumental drift, samples were analyzed in randomized order and each measurement was performed in triplicate.
